```python
import os

f = '1.txt'
f_file = open(f, encoding='utf-8-sig')
puzzle_input = f_file.read()

values = puzzle_input.split('\n')
```

```python
values
```

```python
elf = []
elves = []
for value in values:
    if len(value)>0:
        elf.append(int(value))
    else:
        elves.append(elf)
        elf = []
```

```python
elves
```

```python
elf_totals = []
for elf in elves:
    elf_totals.append(sum(elf))
```

```python
elf_totals
```

```python
max(elf_totals)
```

```python
elf_totals.sort(reverse=True)
```
